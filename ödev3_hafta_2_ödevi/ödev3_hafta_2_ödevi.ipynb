```python
#180401065  1 - Binomial Distribution 
from sympy import Symbol,pprint
import sympy as sym
import sympy.plotting as syp
import matplotlib.pyplot as plt
```

```python
p=Symbol('p')
x=Symbol('x')
n=Symbol('n')
```

```python
my_f_3_part_0=sym.factorial(n)/(sym.factorial(x)*sym.factorial(n-x))
my_f_3_part_0
```

```python

my_f_3_part_1=p**x
my_f_3_part_1
```

```python

my_f_3_part_2=(1-p)**(n-x)
my_f_3_part_2
```

```python
my_f_3=my_f_3_part_0*my_f_3_part_1*my_f_3_part_2
my_f_3
```

```python
sym.plot(my_f_3.subs({p:0.2,n:75}),(x, 0 ,50),title='binomial distribution plot for n=75')
```

```python
x1=[]
y1=[]
for dongu in range(0,51):
    y=my_f_3.subs({p:0.2,n:75,x:dongu}).evalf()
    y1.append(y)
    x1.append(dongu)
plt.plot(x1,y1)
```
